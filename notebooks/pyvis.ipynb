```python
# Parameters
product = "notebooks/pyvis.ipynb"
```

```python
import __path__
```

```python
from config import MAIN_EXPERIMENT


SCALE = MAIN_EXPERIMENT.nb_args[product]
MAIN_MODEL = MAIN_EXPERIMENT.main_model
PICTURES = MAIN_EXPERIMENT.pictures
LDA_FILE = MAIN_EXPERIMENT.lda_file
```

```python
from src.utils import load


lda, dictionary, corpus = load(MAIN_MODEL)
```

```python
import pyLDAvis
import pyLDAvis.gensim


pyLDAvis.enable_notebook()
v = pyLDAvis.gensim.prepare(lda, corpus, dictionary, mds='mmds')
v
```

```python
import json


pyLDAvis.save_json(v, LDA_FILE)
with open(LDA_FILE, 'r', encoding='utf-8') as f:
    data = json.load(f)

with open(LDA_FILE, 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False)
```

```python
import plotly.graph_objects as go

from config import COLORS_A, COLORS_D, COLORS_E

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=data['mdsDat']['x'],
        y=data['mdsDat']['y'],
        mode='text'))

for i, (f, x, y) in enumerate(zip(
        data['mdsDat']['Freq'],
        data['mdsDat']['x'], 
        data['mdsDat']['y'])):
    fig.add_shape(
        line_color=COLORS_E[4],
        fillcolor=COLORS_E[4],
        x0=x-f*SCALE, y0=y-f*SCALE, 
        x1=x+f*SCALE, y1=y+f*SCALE,
        xref='x', yref='y',
        type='circle')

for i, (x, y) in enumerate(zip(
        data['mdsDat']['x'], 
        data['mdsDat']['y'])):
    fig.add_annotation( 
        showarrow=False,
        x=x, y=y,
        text=i)

fig.update_layout(
    font=dict(family='Times New Roman', size=20, color='black'),
    plot_bgcolor='rgba(0, 0, 0, 0)',
    yaxis_scaleanchor="x",
    margin=dict(pad=10),
    height=800, 
    width=800)

fig.update_xaxes(
    zerolinecolor='lightgrey',
    title='PC1',
    gridcolor='lightgrey', 
    showgrid=True, 
    tickangle=0,
    ticklen=10)

fig.update_yaxes(
    zerolinecolor='lightgrey',
    title='PC2',
    gridcolor='lightgrey', 
    showgrid=True, 
    tickangle=0,
    ticklen=10)

fig
```

```python
fig.write_image(f'{PICTURES}/cluster_circles.png')
```
